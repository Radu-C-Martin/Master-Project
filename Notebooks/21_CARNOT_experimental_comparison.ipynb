```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
dfs = []
```

```python
Tsample = 300
```

```python
for idx in range(1, 8):
    dfs.append(pd.read_csv(f"../Data/CARNOT_experimental_validation/Exp{idx}_table.csv").rename(columns = {'Power': 'SimulatedHeat'}))
    dfs[idx-1]['ExperimentTime'] = (dfs[idx-1].index * Tsample) / (24 * 3600)
```

```python
dfs[0]
```

```python
dfs[2][['InsideTemp', 'SimulatedTemp', 'OutsideTemp']].plot(figsize = (15, 5))
```

```python
#fig, (ax1, ax2) = plt.subplots(3, 2, figsize = (20, 5))
fig, (ax1, ax2, ax3) = plt.subplots(3, 2, figsize=(15, 5))

for idx in range(6):
    plt.subplot(3, 2, idx + 1)
    l1 = plt.plot(dfs[idx]['ExperimentTime'], dfs[idx]['OutsideTemp'], ls = '--')
    l2 = plt.plot(dfs[idx]['ExperimentTime'], dfs[idx]['InsideTemp'])
    l3 = plt.plot(dfs[idx]['ExperimentTime'], dfs[idx]['SimulatedTemp'])
    plt.title(f"Experiment {idx + 1}")


labels = ['Measured outside temp', 'Measured inside temp', 'Simulated CARNOT temp']

ax3[0].set_xlabel("Experiment time [days]")
ax3[1].set_xlabel("Experiment time [days]")

ax1[0].set_ylabel("Temperature [°C]")
ax2[0].set_ylabel("Temperature [°C]")
ax3[0].set_ylabel("Temperature [°C]")


#fig.legend([l1, l2, l3], labels=labels, loc = 'upper right')
fig.suptitle("Comparison of CARNOT simulation with real Polydome measurements")

# Adjusting the sub-plots
plt.subplots_adjust(right=0.95)
fig.subplots_adjust(hspace=0.55)

plt.savefig("../Thesis/Plots/CARNOT_comparison_1.pdf", bbox_inches='tight')
```

```python
#fig, (ax1, ax2) = plt.subplots(3, 2, figsize = (20, 5))
fig, ax = plt.subplots(figsize=(15, 3))

idx = 6
l1 = plt.plot(dfs[idx]['ExperimentTime'], dfs[idx]['OutsideTemp'], ls = '--')
l2 = plt.plot(dfs[idx]['ExperimentTime'], dfs[idx]['InsideTemp'])
l3 = plt.plot(dfs[idx]['ExperimentTime'], dfs[idx]['SimulatedTemp'])


labels = ['Measured outside temp', 'Measured inside temp', 'Simulated CARNOT temp']

ax.set_xlabel("Experiment time [days]")

ax.set_ylabel("Temperature [°C]")

fig.suptitle("Experiment 7")

fig.legend([l1, l2, l3], labels=labels, loc = 'upper right')

# Adjusting the sub-plots
plt.subplots_adjust(right=0.95)
fig.subplots_adjust(hspace=0.55)

plt.savefig("../Thesis/Plots/CARNOT_comparison_2.pdf", bbox_inches='tight')
```
